```python
import pandas as pd
import numpy as np
import plotly.express as px

# importing data

article_data = pd.read_csv("articles.leptospirosis.csv")
author_data = pd.read_csv("authors.leptospirosis.csv")
paper_counts = pd.read_csv("paper_counts.csv")
```

```python
author_data['FullName']= author_data['AuthorInitials']+' '+ author_data['AuthorForename']+' '+ author_data['AuthorLastname']
article_data['FullName'] = article_data['FirstAuthorInitials']+' '+ article_data['FirstAuthorForename'] +' '+ article_data ['FirstAuthorLastname'] 
```

```python
def count_top_contributers(count, data):

    article_count = data.groupby('FullName').agg('count')['PMID'].reset_index()

    sorted_article_count = article_count.sort_values(
    by = 'PMID',
    ascending = False
    ).iloc[0:count, :]
    
    return sorted_article_count
```

```python
count_top_contributers(20, article_data)
```

```python

sorted_journal_count = article_data.groupby('Journal').agg('count')['PMID'].reset_index().sort_values('PMID', ascending=False)

sorted_journal_count
```

```python
pivot_journal_count = article_data.value_counts('Journal').reset_index()
journal_counts = pivot_journal_count.rename(columns = {'count': 'Total_Count'})
journal_counts['% '] = round((journal_counts['Total_Count']/journal_counts['Total_Count'].sum())*100, 2)
topt_20 = journal_counts.head(20)
topt_20
```

```python
journal_contribusion = article_data.groupby(['Journal', 'FirstAuthorLastname', 'FullName']).agg('count')['PMID'].reset_index()
journal_contribusion.sort_values('PMID', ascending = False)
```

```python
jouranal_contribution_merged = topt_20.merge(journal_contribusion, how = 'left', on = 'Journal').sort_values(['Total_Count', 'PMID'], ascending = False)
jouranal_contribution_merged
```
